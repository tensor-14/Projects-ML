#Changing Shapes

```python
tensor1=tf.ones([1, 2, 3])  #ONE sublist having TWO sub-sublist having THREE elements
print(tensor1)
```

```python
tensor2=tf.reshape(tensor1, [2, 3, 1])  #TWO sublists having THREE sub-sublists having ONE elements
tensor3=tf.reshape(tensor1, [3, -1])
```

```python
print(tensor2)
print(tensor3)
```
